Você pode executar suas cargas de trabalho primitivas do Qiskit usando REST APIs em um dos três modos de execução, dependendo das suas necessidades: job, session e batch. Este tópico explica esses modos.

> **Note:** Esta documentação utiliza o módulo Python `requests` para demonstrar a REST API do Qiskit Runtime. No entanto, esse fluxo de trabalho pode ser executado usando qualquer linguagem ou framework que suporte o trabalho com REST APIs. Consulte a [documentação de referência da API](https://docs.quantum.ibm.com/api/qiskit-ibm-runtime/tags/jobs) para mais detalhes.

## Modo job com REST API
No modo job, uma única requisição primitiva do Estimator ou do Sampler é feita sem um gerenciador de contexto. Veja como executar um Circuit quântico usando [Estimator](primitives-rest-api#estimator-primitive-with-rest-api) e [Sampler](primitives-rest-api#sampler-primitive-with-rest-api) para alguns exemplos.

## Modo session com REST API
Uma session é um recurso do Qiskit Runtime que permite executar de forma eficiente cargas de trabalho iterativas com múltiplos jobs em computadores quânticos. O uso de sessions ajuda a evitar atrasos causados pelo enfileiramento de cada job separadamente, o que pode ser especialmente útil para tarefas iterativas que exigem comunicação frequente entre recursos clássicos e quânticos. Mais detalhes sobre Sessions podem ser encontrados na [documentação](./execution-modes).

> **Note:** Usuários do Plano Open não podem enviar jobs em session.
